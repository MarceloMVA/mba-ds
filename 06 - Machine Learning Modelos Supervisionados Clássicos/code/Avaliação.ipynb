+ matriz de confusão
<br>A matriz de confusão é uma tabela comparativa dos valores que um algoritmo de machine learning calculou como predição em relação aos valores reais ocorridos e permite extrair métricas que auxiliam na avaliação do modelo.
O nome vem do fato de que torna mais fácil ver se o sistema está confundindo duas classes (ou seja, comumente rotulando uma como outra).
A figura abaixo apresenta, como ilustração, o resultado de previsão de diabetes (SIM/positivo ou NÃO/negativo) para um conjunto de pacientes.

matriz_confusao.jpg
+ verdadeiro positivo
  <br>Quantidade de registros que foram previstos como "positivos" (ou seja, a pessoa foi diagnosticada COM diabetes), e a previsão mostrou-se CORRETA. Conhecido pela sigla TP (true positive). No exemplo acima, equivale ao valor 19.
  
+ verdadeiro negativo 
  <br>Quantidade de registros que foram previstos como "negativos" (ou seja, a pessoa foi diagnosticada SEM diabetes), e a previsão mostrou-se CORRETA. Conhecido pela sigla TN (true negative). No exemplo acima, equivale ao valor 61.

+ falso positivo
  <br>A quantidade de registros que foram previstos como "positivos" (ou seja, a pessoa foi diagnosticada COM diabetes), e a previsão mostrou-se INCORRETA. Conhecido pela sigla FP (false positive). No exemplo acima, equivale ao valor 8.
  
+ falso negativo
  <br>A quantidade de registros que foram previstos como "negativos" (ou seja, a pessoa foi diagnosticada SEM diabetes), e a previsão mostrou-se INCORRETA. Conhecido pela sigla FN (false negative). No exemplo acima, equivale ao valor 13.

+ acurácia
  <br>Indica o quanto o modelo acertou, de forma geral, tanto para previsões negativas quanto para positivas. Fórmula:
  <br><center>$$\frac{( TP + TN )}{( TP + TN + FP + FN )}$$</center>

  <br> No exemplo acima, a acurácia do modelo foi de 79,2% 
  (80 predições corretas [19 + 61] ) / (101 predições totais). 
    
+ precisao
  <br>Indica a proporção de positivos que o modelo realmente acertou nas previsões de positivos, ou seja, dentre aquelas que o modelo indicou como positivas, quantas ele realmente acertou ? Fórmula:
  <br><center>$$\frac{( TP )}{( TP + FP )}$$</center>

  <br> No exemplo acima, a precisão do modelo foi de 70,4%
  (19 predições corretas / (27 ocorrências apontadas como positivas [19 + 8]) 
  
+ recall
	<br>Indica a proporção de positivos que o modelo realmente acertou nas ocorrências de positivos, ou seja, quantos positivos o modelo acertou dentre todos os positivos reais ? Fórmula:
  <br><center>$$\frac{( TP )}{( TP + FN )}$$</center>

  <br> No exemplo acima, o recall do modelo foi de 59,4%
  (19 predições corretas / (32 ocorrências realmente positivas [19 + 13]) 

+ curva ROC
<br>A curva ROC (Receiver Operating Characteristic) é uma ferramenta utilizada na análise de desempenho do modelo. 
A curva é criada plotando-se a taxa de verdadeiros positivos (TPR) em função da taxa de falsos positivos (FPR).
O TPR é a proporção de registros positivos que foram corretamente identificadOs, enquanto o FPR é a proporção de registros negativos que foram erroneamente classificadOs como positivOs. 
Quando o modelo é capaz de distinguir perfeitamente entre instâncias positivas e negativas, a curva ROC atinge o ponto (0,1), ou seja, TPR=1 e FPR=0. 
Já quando o modelo não é capaz de fazer nenhuma distinção entre as classes, a curva ROC coincide com a linha diagonal, 
o que significa que a taxa de verdadeiros positivos é igual à taxa de falsos positivos em todos os pontos de corte.
O objetivo de analisar o poder preditivo do modelo é garantir que ele irá detectar o máximo possível de verdadeiros positivos, enquanto minimiza os falsos positivos (*threshold*).

+ valor AUC
<br>A AUC (area under the ROC curve) é também uma métrica para avaliar o desempenho do modelo, a partir da curva ROC
<br>A AUC corresponde à área total abaixo da curva ROC, que varia entre 0 e 1, sendo que um valor próximo a 1 indica que o modelo é capaz de distinguir corretamente entre instâncias positivas e negativas em diferentes pontos de corte,
 enquanto um valor próximo a 0,5 indica que o modelo não é melhor do que uma escolha aleatória.
Ou seja, a AUC é uma maneira de resumir a curva ROC em um único valor. Quanto maior o AUC, melhor.

```python
#-----------------------------------------------------------------------------------------------------
#3. Criação de um modelo de regressão logística para explicar a variável classe em função das demais.
#-----------------------------------------------------------------------------------------------------

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import ConfusionMatrixDisplay

%notebook matplotlib
from warnings import filterwarnings
filterwarnings('ignore')

# Separando os dados em atributos previsores e classe
X = df_preprocessado.drop('classe', axis=1) # atributos previsores
y = df_preprocessado['classe'] # atributo alvo

# Dividindo o conjunto de dados em treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Treinando o modelo de regressão logística
lrm = LogisticRegression(max_iter=150) #max_iter foi aumentado, pois o valor padrão (100) estava apresentando problemas de convergência
lrm.fit(X_train, y_train)

# Realizando as previsões com o conjunto de teste
y_pred = lrm.predict(X_test)

# Avaliando a acurácia do modelo
print(f'Acurácia do modelo: {(metrics.accuracy_score(y_test, y_pred)*100):.2f}%')

# Exibindo a matriz de confusão
cnf_matrix = metrics.confusion_matrix(y_test,y_pred)
print("Matriz de confusão:")
print(cnf_matrix)

#Apresentação altenativa da matriz de confusão, com layout diferenciado
labels = ['False', 'True']
cm = metrics.confusion_matrix(y_test,y_pred)
ConfusionMatrixDisplay(cm, display_labels=labels).plot()
```
